```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url="https://ekamonline.com/collections/hexa-jars"
```

```python
req=requests.get(url)
```

```python
content=BeautifulSoup(req.content,"html.parser")
```

```python
print(content)
```

```python
name= content.find_all("div",{"class":'grid-product__title'})
price= content.find_all("span",{"class":'sale-price'})
rating= content.find_all("div",{"class":'jdgm-prev-badge'}) 
```

```python
print(name[0])
print(price[0])
print(rating[0])
```

```python
nm=[]
pr=[]
rt=[]
```

```python
for i in name:
    nm.append(i.text)
for i in price:
    pr.append(i.text)
for i in range(len(nm)):
    rt.append(rating[i].text)
```

```python
import pandas as pd
data={"NAME":nm,"PRICE":pr,"RATING":rt}
df=pd.DataFrame(data)
df.transpose()
```

```python
print(df)
```

```python
df.to_csv('AromaticCandles.csv')
```
